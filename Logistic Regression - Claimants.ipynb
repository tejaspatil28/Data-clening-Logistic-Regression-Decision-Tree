## 5. Model Building

```python
X = claimants_data.iloc[:,1:]
y = claimants_data[['ATTORNEY']]
```

```python
X
```

```python
y
```

## K- Fold Cross Validation - Building, Train, Test

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
cv_score = cross_val_score(estimator = log_model, X=X,y=y,cv=5)#Fitting, Training, Test
print('CV score for 5 Fold = {}\nMean Accuracy = {}\nDeviation = {}.'.format(cv_score,cv_score.mean(),cv_score.std()))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 123,test_size = 0.20)
```

```python
X_train
```

```python
X_test
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## 6. Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
#logisticr Regression
log_model=LogisticRegression()
log_model.fit(X_train,y_train)
print(log_model)

#Decision tree classifier
dt_model=DecisionTreeClassifier(criterion='gini',max_depth=8,random_state=18)
dt_model.fit(X_train,y_train)
print(dt_model)

#Random Forest Classifier
rf_model=RandomForestClassifier(n_estimators=120,criterion='gini',max_depth=8,random_state=16)
rf_model.fit(X_train,y_train)
print(rf_model)

#Gradient Boosting Classifier
gb_model=GradientBoostingClassifier(random_state=21)
gb_model.fit(X_train,y_train)
print(gb_model)
```

## 7. Model Testing

#### 7.1 Training data

```python
#Logistic Regression Training prediction
y_pred1_train_data = log_model.predict(X_train)

#Decision Tree training prediction
y_pred2_train_data = dt_model.predict(X_train)

#Random Forest training prediction
y_pred3_train_data=rf_model.predict(X_train)

#Gradient Boosting training prediction
y_pred4_train_data=gb_model.predict(X_train)
```

#### 7.2 Test data

```python
#Logistic regression test prediction
y_pred1_test= log_model.predict(X_test)

#Decision tree testing prediction
y_pred2_test=dt_model.predict(X_test)

#Random Forest testing prediction
y_pred3_test=rf_model.predict(X_test)

#Gradient Boosting prediction
y_pred4_test=gb_model.predict(X_test)
```

## Hyperparameter tuning with GridSearchCV

#### 1. GridSearchCV for DT

```python
from sklearn.model_selection import GridSearchCV
grid_model = GridSearchCV(estimator = dt_model,param_grid = {'criterion' :['gini','entropy'],
                                                             'max_depth' :[3,5,7,8,10], 
                                                             'min_samples_split' :[2,3,4]})
grid_model.fit(X_train,y_train)
print(grid_model.best_params_)
print(grid_model.best_score_)
```

#### 2. GridSearchCV for RandomForest

```python
from sklearn.model_selection import GridSearchCV
grid_model = GridSearchCV(estimator = rf_model,param_grid = {'n_estimators' : [20,50,70,100,150,200],
                                                             'criterion' :['gini','entropy'],
                                                             'max_depth' :[3,5,7,8,10], 
                                                             'min_samples_split' :[2,3,4]})
grid_model.fit(X_train,y_train)
print(grid_model.best_params_)
print(grid_model.best_score_)
```
